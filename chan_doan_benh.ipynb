```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('D:/Pandora Box/dataset benh.csv')
data[6:10]
```

Các giá trị không có gì bất thường.

```python
# Các features đã được xử lý thành categorical variables với 3 mức 'Thấp', 'Trung bình', 'Cao'.

cols_X = ['Danh gia', 'Danh gia.1', 'Danh gia.2', 'Danh gia.3', 'Danh gia.4', 'Danh gia.5']
col_y = ['Chan doan benh']
```

```python
X = data[cols_X] # features
y = data[col_y] # target
```

```python
X.describe()
```

Có thể thấy Danh gia.1 có 1 giá trị ngoài 3 giá trị Thấp, Cao, Trung bình.

```python
X['Danh gia.1'].unique()
```

Do lỗi khi nhập dữ liệu, ta sẽ xử lý lại nó.

```python
X['Danh gia.1'] = pd.DataFrame(['Trung binh' for x in X['Danh gia.1'] if x == 'Trung binh '])

# X['Danh gia.1'] = X['Danh gia.1'].apply(lambda x: 'Trung binh' if x == 'Trung binh ')
# Không thể dùng apply() vì lambda nếu có if sẽ yêu cầu thêm else
```

```python
X.head()
```

```python
# Xử lý các categorical variables thành numerical variables

label_encoder = LabelEncoder()

for col in cols_X:
    X[col] = label_encoder.fit_transform(X[col])
```

```python
model = RandomForestClassifier(n_estimators=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split

# Lấy 90% dữ liệu để train, 10% còn lại để test.

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.1)
model.fit(X_train, y_train)
```

```python
# Dự đoán với 10% dữ liệu đã tách ra ở bên trên

preds = model.predict(X_test)
```

```python
#So sánh kết quả dự đoán với kết quả gốc
from sklearn.metrics import accuracy_score

print(f'Độ chính xác: {accuracy_score(y_test, preds) * 100}%')
```

```python
def chan_doan_benh(array):
    array = np.array(array)
    preds = model.predict(array.reshape(1, -1))
    if preds[0] == 'Gut':
        preds[0] = 'Gút'
    elif preds[0] == 'Tieu duong':
        preds[0] = 'Tiểu Đường'
    elif preds[0] == 'Suy than':
        preds[0] = 'Suy Thận'
    else:
        return 'Bạn không có nguy cơ mắc bệnh.'
    return f'Bạn có nguy cơ mắc bệnh {preds[0]}.'
```
